Se usara la API de COPOMEX (Código Postal Mexicano) para la ubicación, búsqueda y manejo de códigos postales en México para complementar la información de la ubicación por medio del código postal.


```python
import requests
import pandas as pd

pd.set_option('max_rows', 100)
```

Realizamos peticiones de prueba para entender los datos y visualizamos el dataset con la información obtenida.

```python
payload = {'token': '30c1f483-3878-41be-83ba-04c564471848'}
df_ciudad = pd.DataFrame()
```

```python
#Busqueda de los asentamiento (colonia, fraccionamiento, ejido, pueblo) que tienen ese codigo postal

#Codigo postal a buscar
str_cp = '48290'
endpoint = f'https://api.copomex.com/query/info_cp/{str_cp}'

try:
  r = requests.get(endpoint, params=payload)
  if r.status_code == 200:
    list_ciudad = []
    for i in range(1,len(r.json())-1):
      list_ciudad.insert(i, r.json()[i]['response'])
    df_ciudad = pd.DataFrame(list_ciudad)
  else:
    print('Hubo un error en el servicio, Error : ', r.status_code)  
except:
  print('Ocurrió una excepción')  
```
